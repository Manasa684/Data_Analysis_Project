```python
import pandas as pd
import numpy as np
from datetime import datetime
import re
```

```python
df = pd.read_csv('train.csv')
df.head()
```

# Remove the units from some of the attributes and only keep the numerical values (for example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from “New_price”).

```python
# 1. Mileage: remove 'kmpl' and 'km/kg'
df['Mileage'] = pd.to_numeric(
    df['Mileage'].astype(str)
        .str.replace('kmpl', '', regex=False)
        .str.replace('km/kg', '', regex=False)
        .str.strip(),
    errors='coerce'
)

# 2. Engine: remove 'CC'
df['Engine'] = pd.to_numeric(
    df['Engine'].astype(str)
        .str.replace('CC', '', regex=False)
        .str.strip(),
    errors='coerce'
)

# 3. Power: remove 'bhp'
df['Power'] = pd.to_numeric(
    df['Power'].astype(str)
        .str.replace('bhp', '', regex=False)
        .str.strip(),
    errors='coerce'
)
df.head()
```

# Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot encoded value.

```python
# Clean and encode Fuel_Type
df['Fuel_Type'] = df['Fuel_Type'].astype(str).str.strip().str.lower()
df['Fuel_Type'] = df['Fuel_Type'].apply(lambda x: 1 if x == 'petrol' else (0 if x == 'diesel' else None))
df['Fuel_Type'] = df['Fuel_Type'].astype('Int64')

# Clean and encode Transmission
df['Transmission'] = df['Transmission'].astype(str).str.strip().str.lower()
df['Transmission'] = df['Transmission'].apply(lambda x: 1 if x == 'manual' else (0 if x == 'automatic' else None))
df.head()
```

# Creating one more feature and add this column to the dataset (you can use mutate function in R for this). For example, you can calculate the current age of the car by subtracting “Year” value from the current year.

```python
current_year = datetime.now().year  # or simply use 2025 if that's fixed

# Create 'Car_Age' column
df['Car_Age'] = current_year - df['Year']
df.head()
```

# Perform select, filter, rename, mutate, arrange and summarize with group by operations (or their equivalent operations in python) on this dataset.

```python
# SELECT: Selecting a few relevant columns
selected_df = df[['Name', 'Location', 'Year', 'Mileage', 'Price']]
print("\n SELECT Operation Output:")
print(selected_df.head())

# FILTER: Filter cars with price > 10 lakh
filtered_df = df[df['Price'] > 10]
print("\n FILTER Operation Output (Price > 10 Lakh):")
print(filtered_df[['Name', 'Price']].head())

# RENAME: Rename 'Name' column to 'Car_Model'
renamed_df = df.rename(columns={'Name': 'Car_Model'})
print("\n RENAME Operation Output ('Name' → 'Car_Model'):")
print(renamed_df[['Car_Model', 'Location']].head())

# MUTATE: Create a new column 'Price_per_km'
df['Price_per_km'] = (df['Price'] * 100000) / df['Kilometers_Driven']
print("\n MUTATE Operation Output (New column 'Price_per_km'):")
print(df[['Name', 'Kilometers_Driven', 'Price', 'Price_per_km']].head())

# ARRANGE: Sort by 'Price' descending
arranged_df = df.sort_values(by='Price', ascending=False)
print("\n ARRANGE Operation Output (Sorted by 'Price' descending):")
print(arranged_df[['Name', 'Price']].head())
# print(arranged_df.head())

# SUMMARIZE: Average 'Mileage' and 'Price' by original 'Fuel_Type'
# (we group before one-hot encoding to keep it simple)
summary_df = df.groupby('Fuel_Type').agg({'Mileage': 'mean', 'Price': 'mean'}).reset_index()
print("\n SUMMARIZE Operation Output (Grouped by 'Fuel_Type'):")
print(summary_df)
```
