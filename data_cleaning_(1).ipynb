# Look for the missing values in all the columns and either impute them (replace with mean,median, or mode) or drop them. Justify your action for this task.


Dropped New_price because the missing data in the column was more than 5000

```python
# Showing missing values before
print("Missing values before imputation:\n", df.isnull().sum())
```

```python
# Droping columns with more than 30% missing values
missing_ratio = df.isnull().mean()
columns_to_drop = missing_ratio[missing_ratio > 0.3].index
df = df.drop(columns=columns_to_drop)

# Separating object (categorical) and numeric columns
categorical_cols = df.select_dtypes(include='object').columns
numeric_cols = df.select_dtypes(include=['number']).columns

# Imputing categorical columns with mode
df[categorical_cols] = df[categorical_cols].apply(lambda col: col.fillna(col.mode()[0]))

# Imputing numeric columns with median
df[numeric_cols] = df[numeric_cols].apply(lambda col: col.fillna(col.median()))
```

```python
# Showing missing values after
print("\nMissing values after imputation:\n", df.isnull().sum())
df.head()
```
