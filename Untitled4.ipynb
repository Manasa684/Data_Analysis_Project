```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("diabetes.csv")
```

```python
# Set seed and draw sample
np.random.seed(42)
sample_df = df.sample(n=25)

# Calculate statistics
stats_df = pd.DataFrame({
    "Group": ["Population", "Sample (n=25)"],
    "Mean Glucose": [df['Glucose'].mean(), sample_df['Glucose'].mean()],
    "Max Glucose": [df['Glucose'].max(), sample_df['Glucose'].max()]
})

# Plotting
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

sns.barplot(data=stats_df, x="Group", y="Mean Glucose", ax=ax[0], palette="Blues_d")
ax[0].set_title("Mean Glucose Comparison")

sns.barplot(data=stats_df, x="Group", y="Max Glucose", ax=ax[1], palette="Oranges_d")
ax[1].set_title("Max Glucose Comparison")

plt.show()
```

```python
# BMI percentiles
bmi_percentiles = pd.DataFrame({
    "Group": ["Population", "Sample (n=25)"],
    "98th Percentile BMI": [
        np.percentile(df['BMI'], 98),
        np.percentile(sample_df['BMI'], 98)
    ]
})

plt.figure(figsize=(6, 5))
sns.barplot(data=bmi_percentiles, x="Group", y="98th Percentile BMI", palette="Purples_d")
plt.title("98th Percentile of BMI Comparison")
plt.ylabel("BMI")
plt.show()
```

```python
# Bootstrap parameters
n_iterations = 500
sample_size = 150

# Store results
mean_bp = []
std_bp = []
p98_bp = []

np.random.seed(42)
for _ in range(n_iterations):
    bootstrap_sample = df['BloodPressure'].sample(n=sample_size, replace=True)
    mean_bp.append(bootstrap_sample.mean())
    std_bp.append(bootstrap_sample.std())
    p98_bp.append(np.percentile(bootstrap_sample, 98))

# Bootstrap averages
bootstrap_summary = {
    "Metric": ["Mean", "Standard Deviation", "98th Percentile"],
    "Population": [
        df['BloodPressure'].mean(),
        df['BloodPressure'].std(),
        np.percentile(df['BloodPressure'], 98)
    ],
    "Bootstrap Avg (500 samples)": [
        np.mean(mean_bp),
        np.mean(std_bp),
        np.mean(p98_bp)
    ]
}

# Convert to DataFrame
bp_comparison_df = pd.DataFrame(bootstrap_summary)

# Plot
bp_comparison_df.set_index("Metric").plot(kind="bar", figsize=(10, 6), colormap="coolwarm")
plt.title("BloodPressure: Population vs Bootstrap (500 samples)")
plt.ylabel("Blood Pressure Value")
plt.show()
```
